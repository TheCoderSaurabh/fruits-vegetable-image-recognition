```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
ycnn=tf.keras.models.load_model('/content/drive/MyDrive/modelH5/trained_model.h5')
```

```python
test_set = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/VegetableDataset/test',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=32,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)
```

```python
import cv2
image_path='/content/drive/MyDrive/VegetableDataset/test/banana/Image_10.jpg'
img=cv2.imread(image_path)
plt.imshow(img)
plt.title('Test Image')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
image=tf.keras.preprocessing.image.load_img(image_path,target_size=(64,64 ))
input_arr=tf.keras.preprocessing.image.img_to_array(image)
input_arr=np.array([input_arr])
predictions=ycnn.predict(input_arr)
```

```python
print(predictions[0])
print(max(predictions[0]))
```

```python
test_set.class_names
```

```python
result_index=np.where(predictions[0]==max(predictions[0]))
print(result_index[0])
```
